```julia
using MuseInference
using Turing
using Random, LinearAlgebra
using Zygote
Turing.setadbackend(:zygote)
```

# Very simple model

```julia
@model function funnel()
    θ ~ Normal(0, 3)
    z ~ MvNormal(zeros(2048), exp(θ)*I)
    x ~ MvNormal(z, I)
end
```

```julia
Random.seed!(1)
(;x) = rand(funnel() | (θ=0,))
model = funnel() | (;x);
```

```julia
HMC_result = sample(model, HMC(0.1, 10), 1000)
```

```julia
nsims = 145
muse_result = muse(model, 0; nsims, get_covariance=true, progress=true)
```

# LimberJack

```julia
using LimberJack
```

```julia
@model function stat_model()
    θ ~ Normal(0, 3)
    Ωm ~ Uniform(0.1, 0.6)
    Ωb = 0.045 #~ Uniform(0.03, 0.045)
    h  = 0.67 #~ Uniform(0.60, 0.91)
    ns = 0.96 #~ Uniform(0.87, 1.07)
    s8 = 0.811 #~ Uniform(0.6, 0.9)
    
    cosmology = LimberJack.Cosmology(Ωm, Ωb, h, ns, s8,
                                 tk_mode="emulator",
                                 Pk_mode="Halofit");
    ks = LinRange(0.001, 1, 100)
    zs = LinRange(0.001, 2, 50)
    nz = @. (1.4 * exp(-(zs-0.6)^2/0.2))
    tracer = NumberCountsTracer(cosmology, zs, nz, b=1.1)
    cls = angularCℓs(cosmology, tracer, tracer, [0.01, 0.1, 1.0])
    z ~ MvNormal(cls, exp(θ)*I)
    x ~ MvNormal(z, I)
end
```

```julia
@model function stat_model2()
    θ ~ Normal(0, 3)
    Ωm ~ Uniform(0.1, 0.6)
    Ωb = 0.045 #~ Uniform(0.03, 0.045)
    h  = 0.67 #~ Uniform(0.60, 0.91)
    ns = 0.96 #~ Uniform(0.87, 1.07)
    s8 = 0.811 #~ Uniform(0.6, 0.9)
    
    cosmology = LimberJack.Cosmology(Ωm, Ωb, h, ns, s8,
                                 tk_mode="emulator",
                                 Pk_mode="Halofit");
    ks = LinRange(0.001, 1, 100)
    pk = nonlin_Pk(cosmology, ks, 0)
    
    z ~ MvNormal(pk, exp(θ)*I)
    x ~ MvNormal(z, I)
end
```

```julia
Random.seed!(1)
(;x) = rand(stat_model() | (θ=0, Ωm=0.30))
model = stat_model() | (;x);
```

```julia
HMC_result = sample(model, HMC(0.1, 10), 1000)
```

```julia
nsims = 5
muse_result = muse(model, (θ=0, Ωm=0.30); nsims, get_covariance=true, progress=true);
```

# WLWL

```julia
using LimberJack
using PythonCall
using NPZ
```

```julia
np = pyimport("numpy")

fol = "DESY1"
data_set = "wlwl_lite10"
meta = np.load(string("../data/", fol, "/", data_set, "_meta.npz"))
files = npzread(string("../data/", fol, "/", data_set, "_files.npz"))

tracers_names = pyconvert(Vector{String}, meta["tracers"])
pairs = pyconvert(Vector{Vector{String}}, meta["pairs"]);
pairs_ids = pyconvert(Vector{Vector{Int}}, meta["pairs_ids"])
idx = pyconvert(Vector{Int}, meta["idx"])
data_vector = pyconvert(Vector{Float64}, meta["cls"])
cov_tot = pyconvert(Matrix{Float64}, meta["cov"]);

nz_path = "../data/DESY1/lite10_nzs/"
zs_k0, nz_k0, cov_k0 = get_nzs(nz_path, "DESwl__0_e")
zs_k1, nz_k1, cov_k1 = get_nzs(nz_path, "DESwl__1_e")
zs_k2, nz_k2, cov_k2 = get_nzs(nz_path, "DESwl__2_e")
zs_k3, nz_k3, cov_k3 = get_nzs(nz_path, "DESwl__3_e");
```

```julia
@model function stat_model(;
                          tracers_names=tracers_names,
                          pairs=pairs,
                          pairs_id=pairs_ids,
                          idx=idx,
                          cov_tot=cov_tot, 
                          files=files)
    Ωm ~ Uniform(0.1, 0.6)
    Ωb = 0.05 #~ Uniform(0.03, 0.07)
    h = 0.67 #~ Uniform(0.55, 0.91)
    ns = 0.96 #~ Uniform(0.87, 1.07)
    s8 ~ Uniform(0.6, 0.9)
    
    #A_IA ~ Uniform(-5, 5) 
    #alpha_IA ~ Uniform(-5, 5)
    DESwl__0_e_nz ~ MvNormal(nz_k0, cov_k0)
    DESwl__1_e_nz ~ MvNormal(nz_k1, cov_k1)
    DESwl__2_e_nz ~ MvNormal(nz_k2, cov_k2)
    DESwl__3_e_nz ~ MvNormal(nz_k3, cov_k3)
    #DESwl__0_e_m ~ Normal(0.012, 0.023)
    #DESwl__1_e_m ~ Normal(0.012, 0.023)
    #DESwl__2_e_m ~ Normal(0.012, 0.023)
    #DESwl__3_e_m ~ Normal(0.012, 0.023)


    nuisances = Dict("DESwl__0_e_nz" => DESwl__0_e_nz,
                     "DESwl__1_e_nz" => DESwl__1_e_nz,
                     "DESwl__2_e_nz" => DESwl__2_e_nz,
                     "DESwl__3_e_nz" => DESwl__3_e_nz)
    
    cosmology = LimberJack.Cosmology(Ωm, Ωb, h, ns, s8,
                                     tk_mode="emulator",
                                     Pk_mode="linear")
    
    theory = Theory(cosmology, tracers_names, pairs,
                    idx, files;
                    Nuisances=nuisances)
    data_vector ~ MvNormal(theory, cov_tot)
end;
```

```julia
model = stat_model() | (;data_vector);
```

```julia
HMC_result = sample(model, HMC(0.1, 10), 1000)
```

```julia
nsims = 5
muse_result = muse(model, (Ωm=0.30, s8=0.81); nsims, get_covariance=true, progress=true)
```
