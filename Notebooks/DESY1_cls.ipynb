# Plots

```python
data = meta["cls"]
err = np.sqrt(np.diag(meta["cov"]))
pairs = meta["pairs"]
idx = meta["idx"]
cls_EisHu = cls_jaime['DESY1_cls_EisHu']
cls_emul = cls_jaime['DESY1_cls_emul']
```

## DESgc-DESgc

```python
figure, axis = plt.subplots(5, figsize=(5,14))
l = 0
for i in range(0,5):
    t1, t2 = pairs[l][0], pairs[l][1]
    axis[i].errorbar(files["ls_{}_{}".format(t1, t2)], 
                     data[idx[l]:idx[l+1]], 
                     yerr=err[idx[l]:idx[l+1]], fmt='gs', label='Data')
    axis[i].plot(files["ls_{}_{}".format(t1, t2)],
                 cls_EisHu[idx[l]:idx[l+1]], 'bo-', label="EisHu")
    axis[i].plot(files["ls_{}_{}".format(t1, t2)],
                 cls_emul[idx[l]:idx[l+1]], 'r*-', label="Emulator")
    axis[i].set_title("{}_{}".format(t1,t2))
    l += 1
plt.legend()
plt.show()
print(l)
```

## DESgc-DESwl

```python
figure, axis = plt.subplots(5, 4, figsize=(16,16))
l = 5
for i in range(0,5):
    for j in range(0,4):
        t1, t2 = pairs[l][0], pairs[l][1]
        axis[i, j].errorbar(files["ls_{}_{}".format(t1, t2)], 
                     data[idx[l]:idx[l+1]], 
                     yerr=err[idx[l]:idx[l+1]], fmt='gs', label='Data')
        axis[i, j].plot(files["ls_{}_{}".format(t1, t2)],
                     cls_EisHu[idx[l]:idx[l+1]], 'bo-', label="EisHu")
        axis[i, j].plot(files["ls_{}_{}".format(t1, t2)],
                     cls_emul[idx[l]:idx[l+1]], 'r*-', label="Emulator")
        axis[i, j].set_title("{}_{}".format(t1,t2))
        l += 1
plt.legend()
plt.show()
print(l)
```

## DESwl-DESwl

```python
figure, axis = plt.subplots(4, 4, figsize=(16,16))
l = 25
for i in range(0,4):
    for j in range(0,4):
        if j >= i:
            t1, t2 = pairs[l][0], pairs[l][1]
            axis[i, j].errorbar(files["ls_{}_{}".format(t1, t2)], 
                         data[idx[l]:idx[l+1]], 
                         yerr=err[idx[l]:idx[l+1]], fmt='gs', label='Data')
            axis[i, j].plot(files["ls_{}_{}".format(t1, t2)],
                     cls_EisHu[idx[l]:idx[l+1]], 'bo-', label="EisHu")
            axis[i, j].plot(files["ls_{}_{}".format(t1, t2)],
                     cls_emul[idx[l]:idx[l+1]], 'r*-', label="Emulator")
            axis[i, j].set_title("{}_{}".format(t1,t2))
            l += 1
plt.legend()
plt.show()
print(l)
```
